Importando as bibliotecas necessárias. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import requests
import smtplib
from usuario import login, senha
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
```

Criando a conexão com a pagina Web

Acessando o site fundsxplorer e navengando até a aba Rankingig de Fundos Imobiliários

Web Scraping.

Realizando o Web Scraping na tabela Ranking de FII's, em seguida realizando uma leitura da tabela em forma de texto.

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
    (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36'}

url = 'https://www.fundsexplorer.com.br/ranking'
r = requests.get (url, headers= header)

dataframe = pd.read_html(r.text)
print(dataframe)
```

Gerando o DataFrame, da tabela Ranking de FII's

```python
dataframe = pd.DataFrame(dataframe[0])

dataframe
```

Uma breve limpeza na tabela, os dados que não seria necessarios para a ocasião, foram retirados, 
com o objetivo de facilitar a vizualização dos dados mais importantes.

```python
dataframe = dataframe.drop(columns=["DY (3M)Acumulado","DY (6M)Acumulado","DY (3M)Média","DY (6M)Média","DY (12M)Média","DY Ano","Variação Preço","Rentab. Patr.no Período","Rentab.Acumulada","Rentab.Período","VPA","DYPatrimonial","VariaçãoPatrimonial","Rentab. Patr.no Período","Rentab. Patr.Acumulada","VacânciaFinanceira"],axis=1)
dataframe
```

Renomeando as colunas.

```python
df = dataframe.rename(columns={"Preço Atual":"Preço","Liquidez Diária":"Liquidez","DY (12M)Acumulado":"DYAcumulado","PatrimônioLíq.":"Patrimônio","P/VPA":"PVPA"})
df
```

Tratando as colunas e convertendo os dados tipo objetos para float64

```python
df['DividendYield']= df['DividendYield'].apply(lambda x: str(x).replace(",","."))
df['DividendYield']= df['DividendYield'].apply(lambda x: str(x).replace("%",""))
```

```python
df['DividendYield'] = df['DividendYield'].astype('float64')
```

```python
df['DYAcumulado']= df['DYAcumulado'].apply(lambda x: str(x).replace(",","."))
df['DYAcumulado']= df['DYAcumulado'].apply(lambda x: str(x).replace("%",""))
```

```python
df['DYAcumulado']= df['DYAcumulado'].astype('float64')
```

```python
df['Patrimônio'] = df['Patrimônio'].apply(lambda x: str(x).replace(".",""))
df['Patrimônio'] = df['Patrimônio'].apply(lambda x: str(x).replace(",",""))
df['Patrimônio'] = df['Patrimônio'].apply(lambda x: str(x).replace("R$",""))
```

```python
df['Patrimônio']=df['Patrimônio'].astype('float64')
```

```python
df['VacânciaFísica'] = df['VacânciaFísica'].apply(lambda x: str(x).replace(",","."))
df['VacânciaFísica'] = df['VacânciaFísica'].apply(lambda x: str(x).replace("%",""))
```

```python
df['VacânciaFísica'] = df['VacânciaFísica'].astype('float64')
```

FILTROS!  
Para os filtros foi estabelecido um perfil moderado de investidore de acordo com certas características de lidar com o risco. 

Moderado - Aceita um risco maior que o  investidor com perfil cauteloso, em busca de uma rentabilidade maior, mas sempre tentando equilibrar seus investimentos 
entre rentabilidade, liquidez e segurança, através de uma carteira diversificada.

```python
Moderado = df[((df.Liquidez >= 20000.0) & (df.DividendYield >= 0.80) & (df.Patrimônio >= 100000000000) & (df.PVPA <= 90.0)) 
| ((df.Liquidez >= 20000.0) & (df.Patrimônio >= 100000000000) & (df.VacânciaFísica <= 10.00)) & (df.QuantidadeAtivos >= 5)]

Moderado.sort_values(by='Liquidez', ascending=False)
Moderado.to_excel('PERFIL_MODERADO.xlsx',index=False)
Moderado
```

Python e E-mail automático.

```python
msg = MIMEMultipart ()
msg['Subject'] = 'Olá INVESTIDOR(a)! Que tal conferir os FUNDOS IMOBILIÁRIOS para esse mês ?'
msg['From'] = login
msg['To'] = "maarco.eng@outlook.com"
password = senha


corpo_email = """
<p> Olá Investidor(a) , &#128075; </p>

<p>Chegou a hora de conferir as melhores oportunidades em FII's, para esse mês, de acordo com seu perfil de investidor. </p>



<b><FONT SIZE=-1>OS DADOS PRESENTE NESSE E-MAIL NÃO EXCLUI A NECESSIDADE DO ESTUDO E ACOMPANHAMENTO DOS ATIVOS.</FONT></b>
<p> &#9888;&#65039;<font color="#AA0000">Esse e-mail não é uma recomendação de compra ou venda de ativos!</font>

"""

msg.attach(MIMEText(corpo_email, 'html'))
filename1 ="PERFIL_MODERADO.xlsx"



attachment = open(filename1, 'rb')

att = MIMEBase('application','octet-stream')
att.set_payload(attachment.read())
encoders.encode_base64(att)
att.add_header('Content-Disposition', 'attachment; filename = %s' % filename1)



attachment.close()
msg.attach(att)


    

s = smtplib.SMTP('smtp.gmail.com: 587')
s.starttls()

s.login(msg['From'], password)
s.sendmail(msg['From'], [msg['To']],msg.as_string().encode('utf-8'))
print('E_mail enviado')
```
